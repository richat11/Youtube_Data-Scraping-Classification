# Importing Libraries

```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import re
```

```python
# Initializing webdriver session
driver=webdriver.Chrome(executable_path='/Users/vinner/chromedriver')
```

# Fetching top 100 links of each category

```python
# Initializing a list of categories
categories=['Travel','Science','Food','History','Manufacturing','Dance']
```

```python
#base url for any search query in youtube
base_url="https://www.youtube.com/results?search_query="
```

```python
## video_links is an array of arrays containing 100 link of each category
video_links=[]
for i in categories:
    search_url=base_url+i
    driver.get(search_url)
    # Scrolling the webpage multiple times until we get data for 100+ videos
    num=0
    while num<110:
        driver.execute_script('window.scrollBy(0,10000);')
        time.sleep(0.1)
        user_data=driver.find_elements_by_xpath("//*[@id='video-title']")
        num=len(user_data)
    # made an empty list for links and then appending links of the videos to the list
    links=[]
    for i in user_data:
        links.append(i.get_attribute('href'))
    # removing "None"
    links = [i for i in links if i] 
    #taking only 100 youtube links
    links=links[0:100]
    video_links.append(links)
```

```python
len(video_links)
```

# Creating a dataframe, we need to create dataframes for each categories

```python
df = pd.DataFrame(columns = ['link', 'title', 'description', 'category'])
```

# Scraping the video id, title, description and storing them into the dataframe, this is to be done for each category 

```python
wait = WebDriverWait(driver, 10)
j=0
for i in video_links:
    v_category = categories[j]
    j+=1
    for x in i:
            driver.get(x)
            v_id = x.strip('https://www.youtube.com/watch?v=')
            v_title = wait.until(EC.presence_of_element_located(
                           (By.CSS_SELECTOR,"h1.title yt-formatted-string"))).text
            
             # try and except statements are used here so that when Description is not visible then code should not stop 
            
            try:
                v_description = wait.until(EC.presence_of_element_located(
                                     (By.CSS_SELECTOR,"div#description yt-formatted-string"))).text
            except TimeoutException:
                v_description=""
            df.loc[len(df)] = [v_id, v_title, v_description, v_category]
```

# Shuffling the dataframe

```python
df = df.sample(frac=1).reset_index(drop=True)
```

# Storing the data into a csv file

```python
df.to_csv("df_orignal.csv", encoding='utf-8', index=False)
```
